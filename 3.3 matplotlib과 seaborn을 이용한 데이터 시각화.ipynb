```python
# Histogram with Seaborn
fish_data = np.array([2,3,3,4,4,4,4,5,5,6])
fish_data
```

```python
fish_multi = pd.read_csv("./sample/3-3-2-fish_multi_2.csv")
fish_multi
```

```python
fish_multi.groupby("species").describe()
```

```python
length_a = fish_multi.loc[fish_multi["species"] == "A", "length"]
length_b = fish_multi.loc[fish_multi["species"] == "B", "length"]
```

```python
sns.histplot(length_a, bins=5, color="blue", kde=False)
sns.histplot(length_b, bins=5, color="red", kde=False)
```

```python
# Box plot with Seaborn
sns.boxplot(x="species", y="length", data=fish_multi, color="gray")
```

```python
# Violin plot with Seaborn
sns.violinplot(x="species", y="length", data=fish_multi, color="gray")
```

```python
# Bar plot with Seaborn
sns.barplot(x="species", y="length", data=fish_multi, color="gray")
```

```python
# Joint plot with Seaborn
cov_data = pd.read_csv("./sample/3-2-3-cov.csv")
cov_data
```

```python
sns.jointplot(x="x", y="y", data=cov_data, color="black")
```

```python
# Pair plot with Seaborn
iris = sns.load_dataset("iris")
iris.head(3)
```

```python
iris.groupby("species").mean()
```

```python
sns.pairplot(iris, hue="species", palette="gray")
```
